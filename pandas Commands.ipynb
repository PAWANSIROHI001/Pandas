```python
# combine two pandas 

data1 = [10,20,30,40,50,60,70]
data2 = [1,2,3,4,5,6,7]

S1 = pd.Series(data1)
S2 = pd.Series(data2)

S1
S2


def demo(data1, data2):
    if (data1 > data2):
        return data1
    else:
        return data2
    

Result = S1.combine(S2,demo)
print("\n after combine\n",Result)
```

```python
# String  Operations on text data

dataset = { 
         'ID'       :["S01","S02","S03","S04","S05","S06"],
         'students': ["john", "ram", "shyam","shivam","shubham","rishi"],
        'Rank'    :[1,2,3,4,5,6],
        'Marks'   :[98,87,77,69,90,85],
        'Roll No' :[201,202,203,204,205,206]
}      


# create a series

S = pd.Series(dataset)
S

# convert the text to uppercase,lowercase,comelcase,length,count

print(S.str.lower())
print(S.str.upper())
print(S.str.title())
print(S.str.len())
print(S.str.count('john'))
```
